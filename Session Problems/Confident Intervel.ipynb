```python
from scipy.stats import norm
import math

CI = 0.95
Zscore = norm.ppf(1-(1-CI)/2)
SD = 1
n = 30
S_mean = 3.2

ME = Zscore*(SD/math.sqrt(n))

L_bound = S_mean - ME
U_bound = S_mean + ME

print(f"Z-score: {Zscore:.2f}")
print(f"Margin of Error: {ME:.4f}")
print(f"Confidence Interval: ({L_bound:.4f}, {U_bound:.4f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# X values for the normal curve
x = np.linspace(2.0, 4.0, 500)
y = norm.pdf(x, S_mean, SD / np.sqrt(30))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sample Mean Distribution')
plt.fill_between(x, y, where=(x >= L_bound) & (x <= U_bound), color='orange', alpha=0.5, label='95% Confidence Interval')
plt.axvline(S_mean, color='red', linestyle='--', label=f'Sample Mean = {S_mean}')
plt.title("Confidence Interval for Ice Cream Consumption (Known Standard Deviation)")
plt.xlabel("Ice Cream Scoops per Week")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# CI for a Proportion:
import math
n = 200
x = 60
CZ = 1.645
p_hat = x/n
SE = math.sqrt(p_hat*(1-p_hat)/n)
ME = CZ*SE
L_bound = p_hat - ME
U_bound = p_hat +ME
print("CI: ({:.3f}, {:.3f})".format(L_bound, U_bound))
```

```python
# Hypotheses Example:
import scipy.stats as stats
import math 

SM = 78 # sample mean
PM = 80 # population mean
SD = 2.5
SZ = 40 # sample size

Z_statics = (SM - PM) / (SD / math.sqrt(SZ))
alpha = 0.05

CV = stats.norm.ppf(1 - alpha / 2) # critical value of two tailed
rej_n = abs(Z_statics) > CV # reject null value

print(f"Calculated Z-statistic:{Z_statics}")
print(f"Critical Z-value:{CV}")
print(f"Reject the null:{rej_n}")
```

```python
import scipy.stats as stats
import math

CM = 2 # claimed mean
SM = 1.8 # sample mean
S_SD = 0.15 # sample SD
n = 10 # sample size 
alpha = 0.01 # significance level

t_stat = (SM - CM) / ( S_SD / math.sqrt(n))
df = n-1
t_critical = stats. t. ppf(alpha, df)

print(f"t-statistics = {t_stat:.3f}")
print(f"Critical t-value = {t_critical:.3f}")

if t_stat < t_critical:
    print("Reject the null hypothesis: There is enough evidence.")
else:
        print("Fail to reject the null hypothesis: not enough evidence.")
```
