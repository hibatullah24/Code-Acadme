```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Complex dataset
data = [12, 15, 14, 16, 18, 19, 21, 22, 24, 28, 30, 34, 35, 38, 41, 43, 47, 50, 55, 80]

# Statistics
mean = np.mean(data)
std_dev = np.std(data, ddof=1)

# Plot histogram and KDE
plt.figure(figsize=(12, 6))
sns.histplot(data, kde=True, stat="density", color='lightblue', bins=15, label="Data")

# Normal distribution curve
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mean, std_dev), color='blue', linestyle='-', label='Normal Curve')

# Standard deviation lines
for i in range(-3, 4):
    plt.axvline(mean + i*std_dev, color='red' if i == 0 else 'gray', linestyle='--')
    plt.text(mean + i*std_dev, 0.01, f' {"+" if i >= 0 else ""}{i}', rotation=90,
             verticalalignment='bottom', horizontalalignment='center', fontsize=8, color='black')

# Labels
plt.title('Standard Deviation Bell Curve')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm

# Generating three datasets
np.random.seed(0)

# 1. Normal Distribution
normal_data = np.random.normal(50, 10, 1000)  # Mean=50, StdDev=10

# 2. High Skewness (Right Skew)
high_skew_data = np.random.exponential(10, 1000) + 50  # Exponential distribution (right skew)

# 3. Low Skewness (Left Skew)
low_skew_data = -np.random.exponential(10, 1000) + 50  # Negative exponential (left skew)

# Function to calculate measures
def calculate_measures(data):
    mean_value = np.mean(data)
    variance_value = np.var(data, ddof=1)
    std_dev = np.std(data, ddof=1)
    skewness_value = skew(data)
    kurtosis_value = kurtosis(data, fisher=False)
    return mean_value, variance_value, std_dev, skewness_value, kurtosis_value

# Plotting the three distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot Normal Distribution
sns.histplot(normal_data, kde=True, stat="density", ax=axes[0], color='skyblue', bins=30)
axes[0].plot(np.linspace(0, 100, 1000), norm.pdf(np.linspace(0, 100, 1000), np.mean(normal_data), np.std(normal_data)), color='blue', label='Normal Curve')
axes[0].set_title('Normal Distribution')
axes[0].set_xlabel('Values')
axes[0].set_ylabel('Density')

# Plot High Skewness (Right Skew)
sns.histplot(high_skew_data, kde=True, stat="density", ax=axes[1], color='orange', bins=30)
axes[1].set_title('High Skewness (Right)')
axes[1].set_xlabel('Values')
axes[1].set_ylabel('Density')

# Plot Low Skewness (Left Skew)
sns.histplot(low_skew_data, kde=True, stat="density", ax=axes[2], color='green', bins=30)
axes[2].set_title('Low Skewness (Left)')
axes[2].set_xlabel('Values')
axes[2].set_ylabel('Density')

# Calculate and display the measures on each plot
datasets = [normal_data, high_skew_data, low_skew_data]
for i, ax in enumerate(axes):
    mean_value, variance_value, std_dev, skewness_value, kurtosis_value = calculate_measures(datasets[i])
    ax.text(0.05, 0.95, f'Mean: {mean_value:.2f}\nVariance: {variance_value:.2f}\nStd Dev: {std_dev:.2f}\nSkewness: {skewness_value:.2f}\nKurtosis: {kurtosis_value:.2f}',
            transform=ax.transAxes, fontsize=10, verticalalignment='top', horizontalalignment='left', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Generating three datasets
np.random.seed(0)

# 1. Normal Distribution
normal_data = np.random.normal(50, 10, 1000)  # Mean=50, StdDev=10

# 2. High Skewness (Right Skew)
high_skew_data = np.random.exponential(10, 1000) + 50  # Exponential distribution (right skew)

# 3. Low Skewness (Left Skew)
low_skew_data = -np.random.exponential(10, 1000) + 50  # Negative exponential (left skew)

# Calculate measures for the distributions
def calculate_measures(data):
    mean_value = np.mean(data)
    variance_value = np.var(data, ddof=1)
    std_dev = np.std(data, ddof=1)
    skewness_value = skew(data)
    kurtosis_value = kurtosis(data, fisher=False)
    return mean_value, variance_value, std_dev, skewness_value, kurtosis_value

# Calculate and store the measures for the distributions
datasets = [normal_data, high_skew_data, low_skew_data]
labels = ['Normal Distribution', 'High Skewness (Right)', 'Low Skewness (Left)']
measures = []

for i, data in enumerate(datasets):
    measures.append(calculate_measures(data))

# Plotting the distributions in one graph
plt.figure(figsize=(10, 6))

# Colors for each distribution (Shades of Blue)
colors = ['#ADD8E6', '#4682B4', '#1E90FF']  # Light blue, Steel blue, Dodger blue

# Plot Normal Distribution (bell curve)
sns.kdeplot(normal_data, fill=True, color=colors[0], linewidth=2, label='Normal Distribution')

# Plot High Skewness (Right Skew)
sns.kdeplot(high_skew_data, fill=True, color=colors[1], linewidth=2, label='High Skewness (Right)')

# Plot Low Skewness (Left Skew)
sns.kdeplot(low_skew_data, fill=True, color=colors[2], linewidth=2, label='Low Skewness (Left)')

# Adding titles and labels
plt.title('Comparison of Normal, High Skew, and Low Skew Distributions')
plt.xlabel('Values')
plt.ylabel('Density')

# Prepare statistics to display in the legend
stat_legend = []
for i, data in enumerate(datasets):
    mean, variance, std_dev, skewness_value, kurtosis_value = measures[i]
    stat_legend.append(
        f"{labels[i]}:\nMean: {mean:.2f}\nVar: {variance:.2f}\nStd Dev: {std_dev:.2f}\nSkew: {skewness_value:.2f}\nKurt: {kurtosis_value:.2f}"
    )

# Adding the legend with the statistical information
plt.legend(stat_legend, loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.tight_layout()
plt.show()
```
