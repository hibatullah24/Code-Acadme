```python
import numpy as np 
import matplotlib.pyplot as plt

nFlip = 200
C_flip = np.random.randint(0, 2, nFlip)
cum_head = np.cumsum(C_flip)
cum_avg = cum_head/np.arange(1, nFlip+1)

plt.figure(figsize=(10,5))
plt.plot(cum_avg, label='Cumulative Avg of Heads')
plt.axhline(y=0.5, color='red', label='Expected Value(0.5)')
plt.xlabel('Number of Flips')
plt.ylabel('Proportion of Heads')
plt.title('Monte Carlo Simulation of Coin Flips')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

nFlip = 200
T_flip = 3
C_flip = np.random.randint(0, 2, (nFlip,T_flip))
T_head = np.sum(C_flip, axis =1)
cum_head = np.cumsum(T_head)
Total_coins = np.arange(1, nFlip+1)*T_flip
cum_avg = cum_head/Total_coins

plt.figure(figsize=(10,5))
plt.plot(cum_avg, label='Cumulative Avg of Heads')
plt.axhline(y=0.5, color='red', label='Expected Value(0.5)')
plt.xlabel('Number of Flips')
plt.ylabel('Proportion of Heads')
plt.title('Monte Carlo Simulation of Coin Flips')
plt.legend()
plt.grid(True)
plt.show()
```
